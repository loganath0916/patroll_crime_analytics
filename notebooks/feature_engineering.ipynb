```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/crime_cleaned.csv")

print(df.shape)
```

```python
df.columns
```

```python
severity_map = {
    'HOMICIDE': 5,
    'KIDNAPPING': 5,
    'CRIM SEXUAL ASSAULT': 5,
    'ROBBERY': 4,
    'ASSAULT': 4,
    'BATTERY': 3,
    'BURGLARY': 3,
    'MOTOR VEHICLE THEFT': 3,
    'THEFT': 2,
    'NARCOTICS': 2
}

df['Crime_Severity_Score'] = (
    df['Primary Type']
    .map(severity_map)
    .fillna(1)
)
```

```python
df['Lat_Bin'] = pd.cut(
    df['Latitude'],
    bins=20,
    labels=False
)

df['Lon_Bin'] = pd.cut(
    df['Longitude'],
    bins=20,
    labels=False
)
```

```python
from sklearn.preprocessing import LabelEncoder

crime_encoder = LabelEncoder()
location_encoder = LabelEncoder()
district_encoder = LabelEncoder()

df['Crime_Type_Encoded'] = crime_encoder.fit_transform(
    df['Primary Type']
)

df['Location_Encoded'] = location_encoder.fit_transform(
    df['Location Description'].astype(str)
)

df['District_Encoded'] = district_encoder.fit_transform(
    df['District'].astype(str)
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Latitude_Scaled',
    'Longitude_Scaled']] = scaler.fit_transform(
    df[['Latitude','Longitude']]
)
```

```python
new_cols = [
    'Crime_Severity_Score',
    'Lat_Bin',
    'Lon_Bin',
    'Crime_Type_Encoded',
    'Location_Encoded',
    'District_Encoded',
    'Latitude_Scaled',
    'Longitude_Scaled'
]

print(df[new_cols].head())
```

```python
df.to_csv(
    "../data/crime_feature_engineered.csv",
    index=False
)

print("Feature engineering completed")
```
