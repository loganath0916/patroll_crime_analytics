```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv("../data/crime_cleaned.csv")

print("Shape:", df.shape)
df.head()
```

```python
crime_type = (
    df['Primary Type']
    .value_counts()
    .reset_index()
)

crime_type.columns = ['Crime Type', 'Count']

fig = px.bar(
    crime_type.head(15),
    x='Crime Type',
    y='Count',
    title='Top 15 Crime Types',
    color='Count'
)

fig.show()
```

```python
fig = px.scatter_mapbox(
    df.sample(10000),
    lat='Latitude',
    lon='Longitude',
    hover_name='Primary Type',
    zoom=9,
    height=600
)

fig.update_layout(
    mapbox_style='open-street-map'
)

fig.show()
```

```python
hourly = (
    df['Hour']
    .value_counts()
    .sort_index()
    .reset_index()
)

hourly.columns = ['Hour', 'Crime Count']

fig = px.line(
    hourly,
    x='Hour',
    y='Crime Count',
    markers=True,
    title='Crime Distribution by Hour'
)

fig.show()
```

```python
day_order = [
    'Monday','Tuesday','Wednesday',
    'Thursday','Friday',
    'Saturday','Sunday'
]

daily = (
    df['DayOfWeek']
    .value_counts()
    .reindex(day_order)
    .reset_index()
)

daily.columns = ['Day', 'Count']

fig = px.bar(
    daily,
    x='Day',
    y='Count',
    color='Count',
    title='Crime Distribution by Day'
)

fig.show()
```

```python
monthly = (
    df['Month']
    .value_counts()
    .sort_index()
    .reset_index()
)

monthly.columns = ['Month', 'Count']

fig = px.line(
    monthly,
    x='Month',
    y='Count',
    markers=True,
    title='Monthly Crime Trend'
)

fig.show()
```

```python
seasonal = (
    df['Season']
    .value_counts()
    .reset_index()
)

seasonal.columns = ['Season', 'Count']

fig = px.pie(
    seasonal,
    names='Season',
    values='Count',
    title='Seasonal Crime Distribution'
)

fig.show()
```

```python
arrest = (
    df['Arrest']
    .value_counts()
    .reset_index()
)

arrest.columns = ['Arrest', 'Count']

fig = px.pie(
    arrest,
    names='Arrest',
    values='Count',
    title='Arrest Rate Analysis'
)

fig.show()
```

```python
domestic = (
    df['Domestic']
    .value_counts()
    .reset_index()
)

domestic.columns = ['Domestic', 'Count']

fig = px.pie(
    domestic,
    names='Domestic',
    values='Count',
    title='Domestic Crime Analysis'
)

fig.show()
```

```python
district = (
    df['District']
    .value_counts()
    .reset_index()
)

district.columns = ['District', 'Count']

fig = px.bar(
    district,
    x='District',
    y='Count',
    color='Count',
    title='District Wise Crime Distribution'
)

fig.show()
```

```python
cross = pd.crosstab(
    df['Domestic'],
    df['Arrest']
)

fig = px.imshow(
    cross,
    text_auto=True,
    title='Domestic vs Arrest Correlation'
)

fig.show()
```
