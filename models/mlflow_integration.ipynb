```python
import pandas as pd
import mlflow
import mlflow.sklearn

from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering

from sklearn.decomposition import PCA

from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score
)
```

```python
df = pd.read_csv(
    "../data/crime_feature_engineered.csv"
)

print(df.shape)
```

```python
mlflow.set_experiment(
    "Chicago_Crime_Analytics"
)
```

```python
sample_geo = df[
    ['Latitude_Scaled',
     'Longitude_Scaled']
].sample(
    10000,
    random_state=42
)
```

```python
with mlflow.start_run(
    run_name="KMeans_K8"
):

    kmeans = KMeans(
        n_clusters=8,
        init='k-means++',
        n_init=20,
        max_iter=500,
        random_state=42
    )

    labels = kmeans.fit_predict(
        sample_geo
    )

    sil = silhouette_score(
        sample_geo,
        labels
    )

    dbi = davies_bouldin_score(
        sample_geo,
        labels
    )

    mlflow.log_param(
        "algorithm",
        "KMeans"
    )

    mlflow.log_param(
        "n_clusters",
        8
    )

    mlflow.log_metric(
        "silhouette_score",
        sil
    )

    mlflow.log_metric(
        "davies_bouldin",
        dbi
    )

    mlflow.sklearn.log_model(
        kmeans,
        "kmeans_model"
    )

    print("Silhouette:",sil)
    print("DBI:",dbi)
```

```python
with mlflow.start_run(
    run_name="DBSCAN"
):

    dbscan = DBSCAN(
        eps=0.05,
        min_samples=50
    )

    labels = dbscan.fit_predict(
        sample_geo
    )

    sil = silhouette_score(
        sample_geo,
        labels
    )

    dbi = davies_bouldin_score(
        sample_geo,
        labels
    )

    mlflow.log_param(
        "algorithm",
        "DBSCAN"
    )

    mlflow.log_param(
        "eps",
        0.05
    )

    mlflow.log_param(
        "min_samples",
        50
    )

    mlflow.log_metric(
        "silhouette_score",
        sil
    )

    mlflow.log_metric(
        "davies_bouldin",
        dbi
    )

    print("Silhouette:",sil)
    print("DBI:",dbi)
```

```python
with mlflow.start_run(
    run_name="Hierarchical"
):

    hc = AgglomerativeClustering(
        n_clusters=8
    )

    labels = hc.fit_predict(
        sample_geo
    )

    sil = silhouette_score(
        sample_geo,
        labels
    )

    dbi = davies_bouldin_score(
        sample_geo,
        labels
    )

    mlflow.log_param(
        "algorithm",
        "Agglomerative"
    )

    mlflow.log_param(
        "n_clusters",
        8
    )

    mlflow.log_metric(
        "silhouette_score",
        sil
    )

    mlflow.log_metric(
        "davies_bouldin",
        dbi
    )

    print("Silhouette:",sil)
    print("DBI:",dbi)
```

```python
features = [
    'Hour',
    'Month',
    'IsWeekend',
    'Crime_Severity_Score',
    'Crime_Type_Encoded',
    'Location_Encoded',
    'District_Encoded',
    'Latitude_Scaled',
    'Longitude_Scaled',
    'Lat_Bin',
    'Lon_Bin'
]

X = df[features]
```

```python
with mlflow.start_run(
    run_name="PCA"
):

    pca = PCA(
        n_components=3,
        random_state=42
    )

    pca.fit(X)

    variance = (
        pca.explained_variance_ratio_
        .sum()
    )

    mlflow.log_param(
        "components",
        3
    )

    mlflow.log_metric(
        "explained_variance",
        variance
    )

    mlflow.sklearn.log_model(
        pca,
        "pca_model"
    )

    print(
        "Variance:",
        variance
    )
```
